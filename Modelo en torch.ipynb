```python
import torch
import torchaudio
import os
import matplotlib.pyplot as plt
import librosa
```

```python
sig,fs = torchaudio.load("../Audios_DirExam/DirectivityExam00001_ED.wav")# para leer un audio. La se√±al y la frecuencia de muestreo
transform = torchaudio.transforms.Spectrogram(n_fft=1024) # los n_fft se pueden modificar
spectrogram = transform(sig) # Espectrograma
```

```python
sig,fs = torchaudio.load("../Audios_DirExam/DirectivityExam00001_ED.wav")
transform = torchaudio.transforms.MelSpectrogram(fs, n_mels=40) # los n_mels se pueden modificar (40 a 128)
mel_specgram = transform(sig) # Mel-Espectrograma
```

```python
fig, axs = plt.subplots(1, 1)
axs.set_title("Spectrogram (db)")
axs.set_ylabel('bin')
axs.set_xlabel("frame")
im = plt.imshow(librosa.power_to_db(spectrogram[1,:,:]), origin="lower", aspect="auto")
fig.colorbar(im, ax=axs)
```

```python
fig, axs = plt.subplots(1, 1)
axs.set_title("MelSpectrogram (db)")
axs.set_ylabel('bin')
axs.set_xlabel("frame")
im = plt.imshow(librosa.power_to_db(mel_specgram[1,:,:]), origin="lower", aspect="auto")
fig.colorbar(im, ax=axs)
```
